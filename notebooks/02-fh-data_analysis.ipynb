```python
import pandas as pd
import librosa.display
import matplotlib.pyplot as plt
```

```python
data, sr = librosa.load("../symptom-checker.mp3", offset=5, duration=19)
```

```python
import IPython.display as ipd
ipd.Audio(data, rate=sr)
```

```python
plt.figure(figsize=(8, 4))
librosa.display.waveshow(data, sr=sr,  color='#F70000')
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
```

```python
df_test = pd.read_csv("../data/processed/clf/test.csv")
df_train = pd.read_csv("../data/processed/clf/train.csv")
```
